```python
myRawURL_1 = 'https://pkmncards.com/page/' 
myRawURL_2 = '/?s=format%3Astandard&display=images&sort=date&order=asc'

pokemon_entry_urls = []
j = 1
for i in range(1,30):
    myURL = myRawURL_1 + str(i) + myRawURL_2
    uClient = uReq(myURL)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, "html.parser")
    entries = page_soup.find_all("div", {"class": "entry-content"})
    for entry in entries:
        pokemon_entry_urls.append(entry.a['href'])
        print j
        j += 1
```

```python
pokemon_entry_urls[0]
```

```python
folder_path = 'C:\Users\SPECTRE UNIT 01\Desktop\School\Fall 2018\Digital Image Processing\Project\Template\\all_pokemon'

for entry_url in pokemon_entry_urls:
    uClient = uReq(entry_url)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, "html.parser")
    card_img_url = page_soup.find("div", {"class": "card-image"}).a['href']
    file_name = basename(card_img_url)
    file_path = os.path.join(folder_path, file_name)
    urllib.urlretrieve(card_img_url, file_path)
    print 'Saved the image for the card: ', file_name
```

```python
# grabs each row in table
rows = page_soup.findAll("tr")

# Throw away the first row (since it's just the column names)
rows = rows[1:]
```

```python
folder_path = 'C:\Users\SPECTRE UNIT 01\Desktop\School\Fall 2018\Digital Image Processing\Untitled Folder\set_symbols'

for row in rows:
    set_name = row.find("td", {"class":"column-1"}).a.string
    set_image_link = row.find("td", {"class":"column-3"}).img['src']

    file_name = set_name + '.jpg'
    file_path = os.path.join(folder_path, file_name)
    urllib.urlretrieve(set_image_link, file_path)
    print 'Saved the image for the set: ', set_name
```
